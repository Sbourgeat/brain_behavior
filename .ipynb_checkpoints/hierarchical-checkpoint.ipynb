```python
import numpy
import pymc as pm
import arviz as az
import pandas as pd
import os 
import matplotlib.pyplot as plt
```

```python

print(f"Running on PyMC v{pm.__version__}")
```

```python

# Step 1: Load and preprocess the data
data = pd.read_csv("../Results_Vol_Entropy/GWAS_Normalized_EntropyVol.tsv", sep="\t")
#print(data.head(10))

#import behaviour
behav = pd.read_csv("../brain_behavior/formatted_sumamry.csv")
#print(behav.head(10))

# merge data and behav based on DGRP and sex
data = pd.merge(data, behav, on=["DGRP", "sex"])
#print(data.columns)
```

```python

# Step 2: Define the hierarchical model
with pm.Model() as model:
    # Priors for the parameters
    intercept = pm.Normal("intercept", mu=0, sigma=10)
    coef_morpho = pm.Normal("coef_morpho", mu=0, sigma=10)
    coef_micro = pm.Normal("coef_micro", mu=0, sigma=10)
    coef_meso = pm.Normal("coef_meso", mu=0, sigma=10)

    # Step 3: Define the linear regression model
    mu = intercept + coef_morpho * data["Volume"] + coef_micro * data["Entropy0"] + coef_meso * data["Entropy1"]
    sigma = pm.Exponential("sigma", lam=1)  # Prior for the error term

    # Step 4: Define the likelihood
    likelihood = pm.Normal("likelihood", mu=mu, sigma=sigma, observed=data["frac_time_on_shocked"])

    # Step 5: Perform the sampling
    trace = pm.sample(2000, tune=1000, cores=1)  # Use more samples and tune for better results
```

```python
az.plot_trace(trace)
```

```python
az.plot_energy(trace, show=True)
```

```python
az.plot_posterior(trace)
```

```python
az.plot_forest(trace)
```

```python
def split_string_with_dgrp(df, is_sex = True):
    """
    This function takes a dataframe as input and splits the 'genotype' column into two separate lists: DGRP and sex.
    
    Parameters:
    df (DataFrame): The input dataframe containing the 'genotype' and 'sex' columns.
    
    Returns:
    DGRP (list): A list of DGRP values extracted from the 'genotype' column.
    sex (list): A list of sex values extracted from the 'sex' column.
    """
    DGRP=[]
    sex = []
    for i in range(len(df)):
        if 'dgrp' in df.iloc[i, 1]:
            DGRP.append(df.iloc[i, 1])
            if is_sex:
                s= df.iloc[i,2]
                sex.append(s)
    return DGRP,sex
```

```python
def split_string_with_dgrp(df, is_sex = True):
    """
    This function takes a dataframe as input and splits the 'genotype' column into two separate lists: DGRP and sex.
    
    Parameters:
    df (DataFrame): The input dataframe containing the 'genotype' and 'sex' columns.
    
    Returns:
    DGRP (list): A list of DGRP values extracted from the 'genotype' column.
    sex (list): A list of sex values extracted from the 'sex' column.
    """
    DGRP=[]
    sex = []
    for i in range(len(df)):
        if 'dgrp' in df.iloc[i, 1]:
            DGRP.append(df.iloc[i, 1])
            if is_sex:
                s= df.iloc[i,2]
                sex.append(s)
    return DGRP,sex

def merge_data(brain, behav, DGRP, sex=None, is_b2 = False):
    """
    This function merges the brain and behavior data based on the DGRP and sex values.
    
    Parameters:
    brain (DataFrame): The brain data dataframe.
    behav (DataFrame): The behavior data dataframe.
    DGRP (list): A list of DGRP values.
    sex (list): A list of sex values.
    
    Returns:
    merged_df (DataFrame): The merged dataframe containing the brain and behavior data.
    """
    if is_b2:
        data = behav[behav['genotype'].isin(DGRP)]
        data['genotype'] = data['genotype'].apply(lambda x: 'DGRP_0' + x.split('dgrp')[1] if len(x.split('dgrp')[1]) == 2 else 'DGRP_' + x.split('dgrp')[1])

    else:
        data = behav[behav['genotype'].isin(DGRP) & behav["head_scanned"]==True]
        data['genotype'] = data['genotype'].apply(lambda x: 'DGRP_0' + x.split('dgrp')[1] if len(x.split('dgrp')[1]) == 2 else 'DGRP_' + x.split('dgrp')[1])
        #brain['DGRP'] = brain['DGRP'].apply(lambda x: 'DGRP_0' + x if len(x) == 2 else 'DGRP_'+ x)

    
    data.rename(columns={'genotype': 'DGRP'}, inplace=True)


    if sex == None:
        merged_df = pd.merge(brain, data, on='DGRP')
    else:
        merged_df = pd.merge(brain, data, on=['DGRP', 'sex'])
    
    return merged_df
```

```python
behav = pd.read_csv("dgrpool_kl_behaviour.tsv", sep='\t')
brain = pd.read_csv("../Results_Vol_Entropy/GWAS_Normalized_EntropyVol.tsv", sep='\t')
```

```python
data = merged_df
# Step 2: Define the hierarchical model
with pm.Model() as model:
    # Priors for the parameters
    intercept = pm.Normal("intercept", mu=0, sigma=10)
    coef_vol = pm.Normal("coef_vol", mu=0, sigma=10)
    coef_entropy1 = pm.Normal("coef_entropy1", mu=0, sigma=10)
    coef_entropy2 = pm.Normal("coef_entropy2", mu=0, sigma=10)

    # Step 3: Define the linear regression model
    mu = intercept + coef_vol * data["Volume"] + coef_entropy1 * data["Entropy1"] + coef_entropy2 * data["Entropy2"]
    sigma = pm.Exponential("sigma", lam=1)  # Prior for the error term

    # Step 4: Define the likelihood
    likelihood = pm.Normal("likelihood", mu=mu, sigma=sigma, observed=data["KL_frac"])

    # Step 5: Perform the sampling
    trace = pm.sample(2000, tune=1000, cores=-1)  # Use more samples and tune for better results

    idata = trace
    idata.extend(pm.sample_posterior_predictive(idata))


    
```
